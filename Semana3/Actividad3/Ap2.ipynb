# Actividad 3.3
## Antes

```python
df_original = df.copy()

print("\nPrimeras filas:")
print(df.head())

print("\nInformación:")
print(df.info())

print("\nValores nulos:")
print(df.isnull().sum())

print("\nDuplicados:", df.duplicated().sum())
```

## Manejo de valores nulos

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0]) 
```

## Eliminar duplicados

```python
df = df.drop_duplicates()
```

## Convertir tipos de datos

```python
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
```

## Estandarizar nombres de columnas

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")

print("\nColumnas estandarizadas:")
print(df.columns)
```

## Crear nuevas columnas

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1

df['is_alone'] = df['family_size'].apply(lambda x: 1 if x == 1 else 0)
```

## Despues

```python
print("\nPrimeras filas:")
print(df.head())

print("\nInformación:")
print(df.info())

print("\nValores nulos:")
print(df.isnull().sum())

print("\nDuplicados:", df.duplicated().sum())
```

# Actividad 3.4
## Histograma

```python
plt.figure()
plt.hist(df['age'], bins=20)
plt.title("Distribución de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")

plt.savefig("histograma_edad.png")
plt.show()
```

## Grafico de barras

```python
plt.figure()
df['sex'].value_counts().plot(kind='bar')

plt.title("Cantidad de pasajeros por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")

plt.savefig("barras_sexo.png")
plt.show()
```

## Diagrama de dispersion

```python
plt.figure()
plt.scatter(df['age'], df['fare'])

plt.title("Edad vs Tarifa")
plt.xlabel("Edad")
plt.ylabel("Tarifa")

plt.savefig("scatter_edad_tarifa.png")
plt.show()
```

## Mapa de calor de correlaciones

```python
plt.figure()
corr = df.select_dtypes(include=['number']).corr()

sns.heatmap(corr, annot=True)

plt.title("Mapa de calor de correlaciones")

plt.savefig("heatmap_correlaciones.png")
plt.show()
```
